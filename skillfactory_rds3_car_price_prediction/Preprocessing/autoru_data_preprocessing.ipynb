### engine

```python
def get_engine_displacement(text):
    return float(str(text).split('/')[0].split()[0])
```

```python
cars['engine_displacement'] = cars['engine'].apply(get_engine_displacement)
```

```python
def get_engine_power(text):
    
    if len(str(text).split('/')) > 1:
        return int(str(text).split('/')[1].split()[0])
    else:
        return 0
```

```python
cars['engine_power'] = cars['engine'].apply(get_engine_power)
```

```python
def get_fuel(text):
    
    if len(str(text).split('/')) > 1:
        return str(text).split('/')[2]
    else:
        return 0
```

```python
cars['fuel'] = cars['engine'].apply(get_fuel)
```

```python
def mileage_update(row):
    
    return row.split()[0] + row.split()[1]
```

### price

```python
cars.dropna(subset=['price', 'owners', 'vehicle_title'], inplace=True)
```

```python
def price_update(row):   
    
    if len(row.split())<4:
        return row.split()[0] + row.split()[1]
    
    else:
        return row.split()[0] + row.split()[1] + row.split()[2]
```

```python
cars['price'] = cars['price'].apply(price_update)
```

```python
cars['price'] = cars['price'].astype('float64')
```

### owners

```python
cars['owners'] = cars['owners'].apply(lambda x: x.split()[0])
```

```python
cars['owners'] = cars['owners'].astype('int64')
```

```python
cars['features'] = cars['features'].apply(lambda x: 'Unknown' if len(x) == 2 else x)
```

```python
def features_clear(row):
    
    if row == 'Unknown':
        return row
    
    else:
        return row[2:-2].split("', '")
```

```python
cars['features'] = cars['features'].apply(features_clear)
```
