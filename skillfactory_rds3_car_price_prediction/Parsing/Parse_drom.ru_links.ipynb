```python
import csv
import requests
import time

from bs4 import BeautifulSoup
```

```python
def get_page_content(url):
    return requests.get(url).text

def parse_links(page):
    soup = BeautifulSoup(page, 'lxml')
    cars = soup.find_all('a', class_="css-lbefi erw2ohd2")
    
     
    links = []
    
    for car in cars:
        links.append(car.get('href'))
        
    return links
    
```

```python
base_url = 'https://moscow.drom.ru/bmw/used/all/page'

for i in range(1, 80):
    print('Processing page ', i)
    
    url = base_url + str(i) + '/?distance=200'

    content = get_page_content(url)
    links = parse_links(content)
    
    
     
    
    with open('bmw_links_1.txt', 'a') as links_file:
        for link in links:
             links_file.write(link + '\n')
```
